# 3. The command-line interface

Every calculation in this tutorial was expressed as a `Flow` made of
`Work`s made of `Task`s -- the same abstraction whether you're running one
SCF task (Notebook 1) or the hundreds of DFPT tasks behind the phonon flow
(2.4). A few commands worth knowing for automating and inspecting flows
*outside* a notebook, directly from a terminal on the cluster:

```
abirun.py FLOWDIR scheduler      # run (or resume) a flow until completion
abirun.py FLOWDIR status         # check the status of all tasks
abiopen.py FILE                  # open any Abinit netcdf/output file in ipython
abiopen.py FILE -nb              # ... or generate a Jupyter notebook from it
abiopen.py FILE -e               # ... or get a quick summary plot ("expose")
abicomp.py gsr FILE1 FILE2 ...   # compare several files of the same kind
```

`abirun.py FLOWDIR scheduler` is what actually executes a `make_*.py` /
`run_*.py` flow once it's been built and pickled to disk -- every
`flow_*/` directory used in this tutorial was produced this way.

## Robots: the programmatic equivalent of `abicomp.py`

`abicomp.py` compares several files of the same kind from the shell.
AbiPy `Robot`s (`GsrRobot`, `DdbRobot`, ...) do the same thing from Python,
collecting many output files into a single pandas `DataFrame` you can
filter, sort and plot -- this is what powered every convergence study and
comparison plot in `2-Existing_flows`:

```python
with abilab.GsrRobot.from_dir("flow_gaas_convecut") as robot:
    table = robot.get_dataframe()
```

> **Note.** Between the two, prefer `abicomp.py` for a quick look from the
> terminal, and a `Robot` as soon as you need to compute something from the
> collected data (fit an EOS, filter rows, plot a derived quantity, ...).
