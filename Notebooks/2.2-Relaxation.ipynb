# 2.2 Relaxation

*This section is still being written -- check back for updates.*

The idea: rather than fixing the structure and only relaxing the electronic
degrees of freedom (as in every other section of this tutorial), Abinit can
also relax the atomic positions and/or the unit cell itself, driven by the
computed forces and stresses. This is controlled mainly through `ionmov`
(the ionic relaxation algorithm) and `optcell` (whether/how the cell shape
and volume are allowed to change), with convergence controlled by
`tolmxf` (max force) or `tolrff` instead of the `tolvrs`/`tolwfr` used for
a fixed-geometry SCF run.

A worked example (`make_gaas_relax.py` and its analysis) will be added
here.
