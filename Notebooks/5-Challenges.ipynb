# 5.1 Challenges

Time to combine what you've seen. Pick any of the exercises below -- they
don't depend on each other, and none of them has a pre-built `flow_*/`
waiting for you: build, run and analyze each flow yourself, following the
patterns from Notebooks 1-4.

* **A.** A Phonon thermodynamics: Diamond -- ground state, convergence, band
  structure, and (once converged) phonons and their harmonic
  thermodynamics.
* **B.** Phonon convergence study: GaN -- convergence, relaxation, and a quick
  phonon-convergence check.
* **C.** Bring your own material -- download a structure from the
  Materials Project and pseudopotentials from PseudoDojo, then compute
  its electronic band structure.

## A. Phonon thermodynamics: Diamond

`../Data/Structures/mp-66_C.cif` and `../Data/Pseudos/C.psp8` are already
in place. Using the patterns from Notebook 1 (Task/Flow basics) and
Notebook 2 (`wlib.build_ecut_conv_flow` for the convergence study,
`wlib.build_gaas_ebands_flow` for the band structure), build (some or all
of): a ground-state run, an ecut convergence study, and a band structure
for diamond.

Diamond is face-centered cubic, like GaAs and Si, so the same L-Gamma-X
k-path (`wlib.FCC_KPATH`) applies to its band structure.

Once you've settled on a converged `ecut`, go one step further: compute
diamond's phonons and their harmonic thermodynamics. Build a `PhononFlow`
for diamond (`wlib.build_mgo_phonon_flow` as a template -- swap in
diamond's structure/pseudo and your converged `ecut`), post-process it
with `anaddb` as in `4-Phonons.ipynb`, then get the resulting `PhononDos`'s
`get_cv`/`get_entropy`/`get_free_energy`/`get_internal_energy` (or
`plot_harmonic_thermo()` for all four at once) -- but only *after* the
convergence study above; a phonon calculation is only as good as the
`ecut` (and k-/q-mesh) it's built on.

## B. Phonon convergence study: GaN

`../Data/Structures/mp-804_GaN.cif` and the `Ga.psp8`/`N.psp8` pseudos are
already in place -- GaN (wurtzite) is isostructural with the AlN from
Notebook 3, so its relaxation input can follow `wlib.aln_relax_input`
almost line for line.

1. **Convergence.** Following `wlib.gs_input`/`wlib.build_ecut_conv_flow`
   (Notebook 2) as a template, write your own GaN ground-state input and
   ecut convergence flow. `wlib.plot_ecut_conv` works on any flow with one
   SCF task per `ecut`, regardless of the material.
2. **Relaxation.** Following `wlib.aln_relax_input`/
   `wlib.build_aln_relax_flow` (Notebook 3), relax GaN's atomic positions
   and cell volume. Wurtzite has two internal degrees of freedom (the
   `c/a` ratio and the internal `u` parameter) on top of the volume --
   does `optcell=1` (volume only) relax enough of them, or would you need
   `optcell=2` (full cell, still symmetry-constrained)?
3. **Phonons at Gamma vs. ecut.** Before committing to a full (coarse)
   `PhononFlow` like `wlib.build_mgo_phonon_flow` (Notebook 4), it's worth
   checking that the *cheapest* possible phonon calculation -- just the
   `q = Gamma` point, i.e. `ph_ngqpt=(1, 1, 1)` -- is itself converged
   with `ecut`. Build one `PhononFlow.from_scf_input(..., ph_ngqpt=(1, 1, 1))`
   per `ecut` value (same loop structure as step 1's ecut convergence),
   then compare the Gamma-point phonon frequencies across `ecut` --
   `ddb.anaget_phmodes_at_qpoint()` (default `qpoint` is Gamma) gives you a
   `PhononBands` object for a single q-point without needing a full
   `anaddb` interpolation.

## C. Bring your own material

Everything so far has used structures and pseudopotentials already placed
in `../Data/`. This time, get your own:

1. **Structure.** Pick any material on the
   [Materials Project](https://materialsproject.org/) website (free
   account) and download its CIF file -- same format as the `mp-*.cif`
   files already in `../Data/Structures/`. Save it there (or point
   `abilab.Structure.from_file` at wherever you saved it).
2. **Pseudopotentials.** For each element in your structure, download the
   matching norm-conserving pseudopotential from
   [PseudoDojo](http://www.pseudo-dojo.org/) -- same table
   (norm-conserving, scalar-relativistic, PBE) as the `.psp8` files
   already in `../Data/Pseudos/`.
3. **Band structure.** Following the pattern from Notebook 2
   (`wlib.build_gaas_ebands_flow`, or `_bandstructure_inputs` +
   `flowtk.bandstructure_flow` directly), build and run a band structure
   flow for your material. `wlib.FCC_KPATH` only applies to cubic (fcc)
   structures like GaAs/Si/diamond -- for anything else, AbiPy can work
   out a high-symmetry k-path from your structure's own symmetry:
   `[kpt.frac_coords for kpt in structure.hsym_kpoints]` gives a
   `kptbounds` list in the same format as `wlib.FCC_KPATH`.
