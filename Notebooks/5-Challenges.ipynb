## C. Phonon thermodynamics

`phdos_file.phdos`, the phonon density of states of MgO. Within the
harmonic approximation, a phonon DOS is enough to get thermodynamic
properties as a function of temperature (internal energy, entropy, free
energy, heat capacity).

Re-open the phonon DOS file if you closed it:

